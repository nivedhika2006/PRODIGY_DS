```python
# ============================================================
# PRODIGY INFOTECH - DATA SCIENCE INTERNSHIP
# Task 02: Data Cleaning & Exploratory Data Analysis (EDA)
# Dataset: Titanic Dataset
# ============================================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ── Load Dataset ──────────────────────────────────────────
df = pd.read_csv('/content/train.csv')  # change filename if different

print("Shape:", df.shape)
print("\nFirst 5 rows:")
print(df.head())

# ── Data Cleaning ─────────────────────────────────────────
print("\nMissing Values:")
print(df.isnull().sum())

# Fill missing Age with median
df['Age'] = df['Age'].fillna(df['Age'].median()) # Updated to remove inplace=True

# Fill missing Embarked with mode
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0]) # Updated to remove inplace=True

# Drop Cabin (too many missing)
df.drop(columns=['Cabin'], inplace=True)

print("\nAfter Cleaning - Missing Values:")
print(df.isnull().sum())

# ── Chart 1: Survival Count ───────────────────────────────
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', data=df, hue='Survived', palette='Set2', legend=False) # Updated to address palette warning
plt.title('Survival Count', fontsize=14)
plt.xticks([0, 1], ['Did Not Survive', 'Survived'])
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('Task02_survival_count.png', dpi=150)
plt.show()

# ── Chart 2: Survival by Gender ───────────────────────────
plt.figure(figsize=(6, 4))
sns.countplot(x='Sex', hue='Survived', data=df, palette='Set1')
plt.title('Survival by Gender', fontsize=14)
plt.ylabel('Count')
plt.legend(['Did Not Survive', 'Survived'])
plt.tight_layout()
plt.savefig('Task02_survival_by_gender.png', dpi=150)
plt.show()

# ── Chart 3: Survival by Passenger Class ──────────────────
plt.figure(figsize=(6, 4))
sns.countplot(x='Pclass', hue='Survived', data=df, palette='coolwarm')
plt.title('Survival by Passenger Class', fontsize=14)
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(['Did Not Survive', 'Survived'])
plt.tight_layout()
plt.savefig('Task02_survival_by_class.png', dpi=150)
plt.show()

# ── Chart 4: Age Distribution ─────────────────────────────
plt.figure(figsize=(8, 4))
sns.histplot(df['Age'], bins=30, kde=True, color='steelblue')
plt.title('Age Distribution of Passengers', fontsize=14)
plt.xlabel('Age')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('Task02_age_distribution.png', dpi=150)
plt.show()

# ── Chart 5: Correlation Heatmap ──────────────────────────
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap', fontsize=14)
plt.tight_layout()
plt.savefig('Task02_heatmap.png', dpi=150)
plt.show()

print("✅ All Task 02 charts saved!")
```
